```python
from __future__ import print_function
import nltk
import textacy

from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
from nltk.stem.lancaster import LancasterStemmer

class NLPUtils(object):
    def __init__(self):
        self.st = LancasterStemmer()

    def get_ngrams(self, text, n):
        tokens = nltk.word_tokenize(text)
        return nltk.ngrams(tokens, n)

    def ngram_stats(self, text, n):
        ngrams = self.get_ngrams(text, n)
        fdist = nltk.FreqDist(ngrams)
        res = {}
        for k,cnt in fdist.items():
            key = " ".join(k)
            res[key] = cnt

        return res, sorted(list(res.items()), key=lambda x: x[1], reverse=True)

    def get_txt_ngrams_from_spcy_doc(self, doc, n):
        ngram_spans = textacy.extract.ngrams(doc, n)

    def ngram_stats_from_spcy_doc(self, doc, n):
        ngrams = self.get_txt_ngrams_from_spcy_doc(text, n)

    def get_setences(self, text):
       sent_tokenize_list = sent_tokenize(text)
       return [x for x in sent_tokenize_list if not x == "\n"] # Makign sure there is no new empty line as a sentence

    def get_stems(self, text):
        """ Get stems for the original words in the text using Lanchester Stemmer via nltk """
        # Assuming the text is normalized - each word separated by space and lower-cased and each word
        text_stem_dict= {}

        for word in text.split():
            text_stem_dict[word] = self.st.stem(word)

        return text_stem_dict

    def get_stop_words(self, language='english'):
        """ Returns a set of stop words of a given language. default: english """
        return set(stopwords.words(language))

    def rem_ws_spacy_span(self, spacy_span):
        """
        remove leading and trailing whitespace from spacy span.
        This includes cases where a span starts with \n
        :param spacy_span:
        :return: spacy span just like the input but with leading and trailing space tokens removed. if the span consits
        of all space tokens, None is returned.
        """

        # return None if the span is all whitespace. TODO check if empty span is possible, if so return that instead.
        if all([t.is_space for t in spacy_span]):
            return None

        # get number of leading space type tokens
        num_leading = 0
        for t in spacy_span:
            if t.is_space:
                num_leading += 1
            else:
                break

        # get number of trailing space type tokens
        num_trailing = 0
        for t in reversed(spacy_span):
            if t.is_space:
                num_trailing += 1
            else:
                break

        # This should be impossible. Would occur if all tokens were white space, but this should have been checked above
        # assert num_leading + num_trailing < len(spacy_span) TODO better to log as error and keep going

        # determine start and end index for slice of span which will eliminate leading and trailing whitespace tokens
        start = num_leading
        end = len(spacy_span) - num_trailing
        new_span = spacy_span[start:end]

        return new_span

    def get_space_norm_text_and_map(self, input_string):
        norm_string = ""
        index_map = list()  # maps normalized text index to original text index
        prev_was_space = False
        for i, c in enumerate(input_string):
            if c.isspace() or c == '\r':
                if prev_was_space is True:
                    continue
                else:
                    prev_was_space = True
                    norm_string += c
                    index_map.append(i)
            else:
                norm_string += c
                index_map.append(i)
                prev_was_space = False

        return norm_string, index_map

    def spacy_token_list_to_char_offset_pair(self, token_list):
        start = token_list[0].idx
        end = token_list[-1].idx + len(token_list[-1].text)  # NOTE: this works if end offset is meant to be one past the index of the last character.

        return (start, end)

    # TODO: following function is temporarily useful. we should use the normalization function below this one (lemma based) when computing IDF scores in the future, in which case we can use it in place of this one.
    # ^^ @ankit note this when creating data processing pipeline! we can discuss at that time. - Taylor
    def spacy_token_list_to_lower_string(self, token_list):
        """
        This normalization is used to compute tfidf. we don't lemmatize since lemmatization was not done when processing
        enron corpus.


        each token is converted to lowercase; whitespace preserved except for after last token
        :param token_list: list of spacy tokens, or a spacy span
        :return:
        """
        lower_string = ' '.join([t.lower_ for t in token_list])
        return lower_string

    def spacy_token_list_to_lemma_string(self, token_list):
        """
        This method converts a spacy token list to a string in the normalized form that is used for:
        1. display in discovered topics pane
        2. key in dicts or sets used to group phrases that are considered equivalent.

        Currently the normalization process is:
        lemmatize and lowercase each token string.
        separate each pair of consecutive tokens by
        (i) a single whitespace, if they are separated by any amount or type of whitespace in the text, or
        (ii) no whitespace or separation character at all (i.e. empty string)

        TODO pronouns should not be converted to -pron-, the string should be kept!

        :param token_list: list of spacy tokens, or a spacy span
        :return:
        """
        _lemmatized_string = list()
        for t in token_list[:-1]:
            if len(t.text_with_ws) > len(t.text):
                whitespace = True
            else:
                whitespace = False

            _lemmatized_string.append(self.spacy_token_to_lemma_lower(t))
            if whitespace is True:
                _lemmatized_string.append(' ')

        t = token_list[-1]
        _lemmatized_string.append(t.lemma_)

        lemmatized_string = ''.join(_lemmatized_string)

        return lemmatized_string

    def spacy_token_to_lemma_lower(self, t):
        return t.lemma_.lower().strip()

    def spacy_tok_is_adj(self, t):
        if t.pos_ == "ADJ":
            return True
        else:
            return False

    def spacy_tok_is_pron(self, t):
        if t.pos_ == "PRON":
            return True
        else:
            return False
```

```python
""" Base class for Spacy /Textacy Extractor """
from __future__ import absolute_import, division, print_function, unicode_literals
import textacy

"""
NOTICE: this file is the modified version of *** from the textacy library.
this notice is here per Apache 2.0 license.
"""

import spacy

"""
Functions for unsupervised automatic key term extraction, both specific algorithms
(SGRank, TextRank, SingleRank) and a generalization of semantic network-based methods.
Also includes a function to aggregate common key term variants of the same idea.
"""


from collections import defaultdict
import itertools
from operator import itemgetter
import networkx as nx

from textacy.network import terms_to_semantic_network
from textacy.keyterms import rank_nodes_by_divrank
from textacy.keyterms import rank_nodes_by_bestcoverage


class TextacyExtractor():
    def __init__(self, nlp):
        self.nlp = nlp # have to download these seperately
        self.nlp_utils = NLPUtils()
        # TODO the spacy_lang will be attribute of KeywordExtractor, but in any case,
        # note that if one doesn't have en_default installed key_terms_from_semantic_network may not function properly

    def key_terms_from_semantic_network(self, doc, window_width=3, edge_weighting='cooc_freq', ranking_algo='pagerank',
                                        join_key_words=False, n_keyterms=10, **kwargs):
        """
        Extract key terms from a document by ranking nodes in a semantic network of
        terms, connected by edges and weights specified by parameters.

        Args:
            doc (``textacy.Doc`` or ``spacy.Doc``)
            normalize (str or callable): if 'lemma', lemmatize terms; if 'lower',
                lowercase terms; if None, use the form of terms as they appeared in
                ``doc``; if a callable, must accept a ``spacy.Token`` and return a str,
                e.g. :func:`textacy.spacy_utils.normalized_str()`
            window_width (int): width of sliding window in which term
                co-occurrences are said to occur
            edge_weighting ('binary', 'cooc_freq'}): method used to
                determine weights of edges between nodes in the semantic network;
                if 'binary', edge weight is set to 1 for any two terms co-occurring
                within `window_width` terms; if 'cooc_freq', edge weight is set to
                the number of times that any two terms co-occur
            ranking_algo ({'pagerank', 'divrank', 'bestcoverage'}):
                algorithm with which to rank nodes in the semantic network;
                `pagerank` is the canonical (and default) algorithm, but it prioritizes
                node centrality at the expense of node diversity; the other two
                attempt to balance centrality with diversity
            join_key_words (bool): if True, join consecutive key words
                together into longer key terms, taking the sum of the constituent words'
                scores as the joined key term's combined score
            n_keyterms (int or float): if int, number of top-ranked terms
                to return as keyterms; if float, must be in the open interval (0, 1),
                is converted to an integer by ``round(len(doc) * n_keyterms)``

        Returns:
            List[Tuple[str, float]]: sorted list of top ``n_keyterms`` key terms and
                their corresponding ranking scores

        Raises:
            ValueError: if ``n_keyterms`` is a float but not in (0.0, 1.0]
        """

        #def get_norm(token):
         #   if normalize == 'lower':
          #      return token.lower_
          #  elif normalize == 'lemma':
          #      return token.lemma_.lower()
          #  elif not normalize:
          #      return token.text
          #  else:
          #      return normalize(token)

        if isinstance(n_keyterms, float):
            if not 0.0 < n_keyterms <= 1.0:
                raise ValueError('`n_keyterms` must be an int, or a float between 0.0 and 1.0')
            n_keyterms = int(round(len(doc) * n_keyterms))

        phrase_to_token_list = defaultdict(list)  # map string to list of lists of tokens. TODO should list of tokens be span instead?
        include_pos = {'NOUN', 'PROPN', 'ADJ'}
        word_list = list()
        token_list = list()

        for word in doc:
            word_list.append(self.nlp_utils.spacy_token_to_lemma_lower(word))
            token_list.append(word)
            phrase_to_token_list[self.nlp_utils.spacy_token_to_lemma_lower(word)].append([word])  # TODO: probably better to get this mapping using same
                                                                # filtering applied below to build good_word_list

        good_word_list = [self.nlp_utils.spacy_token_to_lemma_lower(word) for word in doc
                          if not word.is_stop and not word.is_punct and
                          word.pos_ in include_pos and word.text.strip()] #  "and word.text.strip()" means to get rid of
                                                                            # sequences of tabs and line breaks

        if not good_word_list:
            return list(), phrase_to_token_list  # phrase2tokenlit will be empty defaultdict(list)
        graph = terms_to_semantic_network(
            good_word_list, window_width=window_width, edge_weighting=edge_weighting)

        # rank nodes by algorithm, and sort in descending order
        word_ranks = None
        if ranking_algo == 'pagerank':
            word_ranks = nx.pagerank_scipy(graph, weight='weight')
        elif ranking_algo == 'divrank':
            word_ranks = rank_nodes_by_divrank(
                graph, r=None, lambda_=kwargs.get('lambda_', 0.5), alpha=kwargs.get('alpha', 0.5))
        elif ranking_algo == 'bestcoverage':
            word_ranks = rank_nodes_by_bestcoverage(
                graph, k=n_keyterms, c=kwargs.get('c', 1), alpha=kwargs.get('alpha', 1.0))

        # bail out here if all we wanted was key *words* and not *terms*
        if join_key_words is False:
            return [(word, score) for word, score in
                    sorted(word_ranks.items(), key=itemgetter(1), reverse=True)[:n_keyterms]], phrase_to_token_list

        top_n = int(0.90 * len(word_ranks))
        top_word_ranks = {word: rank for word, rank in
                          sorted(word_ranks.items(), key=itemgetter(1), reverse=True)[:top_n]}

        # join consecutive key words into key terms
        seen_joined_key_terms = set()
        joined_key_terms = []
        for key, group in itertools.groupby(token_list,
                                            lambda t: self.nlp_utils.spacy_token_to_lemma_lower(t) in top_word_ranks):
            if key is True:
                words = list(group)
                norm_term = self.nlp_utils.spacy_token_list_to_lemma_string(words)

                phrase_to_token_list[norm_term].append(words)  # term --> [token1, ... tokenk] for term of length k

                if norm_term in seen_joined_key_terms:
                    continue

                seen_joined_key_terms.add(norm_term)
                joined_key_terms.append((norm_term,
                                         sum(word_ranks[self.nlp_utils.spacy_token_to_lemma_lower(w)] for w in words)))

        return sorted(joined_key_terms, key=itemgetter(1, 0), reverse=True)[:n_keyterms], phrase_to_token_list

    def extract(self, text_or_doc):
        # get matched phrases
        if type(text_or_doc) is not spacy.tokens.doc.Doc and type(text_or_doc) is not textacy.Doc:
            doc = self.nlp(text_or_doc)
        else:
            doc = text_or_doc

        info, mapping = self.key_terms_from_semantic_network(doc, join_key_words=True, n_keyterms=.50)

        if info is None:
            info = []

        return info, mapping


# if __name__ == "__main__":
#     te = TextacyExtractor(spacy.load('en'))
#     text = u"Climate change is a huge issue around the world. \n We have agreed on reducing the carbon footprint across " \
#            u"all our industries.\n Governments have to stand with the environment regulators to ensure these agreements are fulfilled."
#     print(te.extract(text))
```
